<h1>Comparing the time complexity of two solutions to the same problem</h1>

The goal was to write a program to print the nth Fibonacci number. The first solution, fib1, does so recursively and has exponential time complexity. The second solution, fib2, uses an array to eliminate the repetitive calculations done in fib1, allowing us to achieve O(n) complexity. The third solution, fib3, solves the recurrence relation F(n) = F(n-1) + F(n-2), in order to achieve log(n) complexity (due to the power function having O(log(n)) complexity)

<b>Let's compare the functions by plotting the relationship between the size of n and execution time in our two functions!</b>

```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def fib1(n):
    """
    Recursive solution with O(2^n)
    """
    if n<=1:
        return n
    return fib1(n-1)+fib1(n-2)

def fib2(n):
    """
    Use an array to achieve O(n)
    """
    f = [0, 1]
     
    for i in range(2, n+1):
        f.append(f[i-1] + f[i-2])
    return f[n]

def fib3(n):
    """
    Solved recurrence relation with O(log(n))
    """
    num = ((1+np.sqrt(5)) / 2) ** n
    return round(num / np.sqrt(5))

def main():
    arr = np.arange(0, 38, 2)
    y = []
    for i in arr:
        start_time = time.perf_counter()
        fib1(i)
        y.append(time.perf_counter() - start_time)
       
    z = []
    for j in arr:
        start_time = time.perf_counter()
        fib2(j)
        z.append(time.perf_counter() - start_time)
        
    a = []
    for k in arr:
        start_time = time.perf_counter()
        fib3(k)
        a.append(time.perf_counter() - start_time)
        
    # plot execution times
    plt.figure(figsize = (15,15)) 
    plt.title("Execution time of fib1 vs fib2 vs fib3", fontsize=20)
    plt.xlabel("Values of N", fontsize=15) 
    plt.ylabel("Time to Execute (seconds)", fontsize=15) 
    plt.plot(arr, y, label="fib1 - Recursively acquired F(n)")
    plt.plot(arr, z, label="fib2 - Used an array to solve for F(n)")
    plt.plot(arr, a, label="fib3 - Solved recurrence relation for F(n)")
    plt.legend(fontsize=15, bbox_to_anchor=(1.48, 1.00))
    plt.show()
    
main()
```

<h4>As we can see, fib2 and fib3 perform far better than fib1 in terms of execution time! But Who performs best out of fib2 and fib3?</h4>

```python
def compare2():
    arr = np.arange(0, 100, 1)
       
    z = []
    for j in arr:
        start_time = time.perf_counter()
        fib2(j)
        z.append(time.perf_counter() - start_time)
        
    a = []
    for k in arr:
        start_time = time.perf_counter()
        fib3(k)
        a.append(time.perf_counter() - start_time)
        
    # plot execution times
    plt.figure(figsize = (15,15)) 
    plt.title("Execution time of fib2 vs fib3", fontsize=20)
    plt.xlabel("Values of N", fontsize=15) 
    plt.ylabel("Time to Execute (seconds)", fontsize=15) 
    plt.plot(arr, z, label="fib2 - Used an array to solve for F(n)")
    plt.plot(arr, a, label="fib3 - Solved recurrence relation for F(n)")
    plt.legend(fontsize=15, bbox_to_anchor=(1.48, 1.00))
    plt.show()

compare2()
```

<b>Initially, fib2 is faster than fib3. However, after n reaches a certain size, fib3 becomes faster!</b>
