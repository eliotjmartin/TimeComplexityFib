<h1>Comparing the time complexity of two solutions to the same problem</h1>

The goal was to write a program to print the nth Fibonacci number. The first solution, fib1, does so recursively and has exponential time complexity. The second solution, fib2, solves the recurrence relation F(n) = F(n-1) + F(n-2), in order to achieve log(n) complexity (due to the power function having O(log(n)) complexity)

<b>Let's compare the solutions execution time by plotting the relationship between the size of n and execution time in our two functions!</b>

```python
import time
import numpy as np
import matplotlib.pyplot as plt

def fib1(n):
    """
    Recursive solution with O(2^n)
    """
    if n<=1:
        return n
    return fib1(n-1)+fib1(n-2)

def fib2(n):
    """
    Solved recurrence relation with O(log(n))
    """
    num = ((1+np.sqrt(5)) / 2) ** n
    return round(num / np.sqrt(5))

def main():
    arr = np.arange(0, 38, 2)
    y = []
    for i in arr:
        start_time = time.time()
        fib1(i)
        y.append(time.time() - start_time)
       
    z = []
    for j in arr:
        start_time = time.time()
        fib2(i)
        z.append(time.time() - start_time)
        
    # plot execution times
    plt.figure(figsize = (10,10)) 
    plt.title("Execution time of fib1 vs fib2", fontsize=20)
    plt.xlabel("Values of N", fontsize=15) 
    plt.ylabel("Time to Execute (seconds)", fontsize=15) 
    plt.plot(arr, y, label="Recursively acquired F(n)")
    plt.plot(arr, z, label="Solved for F(n)")
    plt.legend(fontsize=15, bbox_to_anchor=(1.48, 1.00))
    plt.show()
    
main()
```

<h4>As we can see, fib2 performs far better than fib1 in terms of execution time!</h4>
